```python
import pandas as pd
import numpy as np
import json
df=pd.read_csv("line_details.csv")
# delete the second and third column
df.drop(df.columns[[5,6,8,9,10,11,12]], axis=1, inplace=True)
# df['id']=[i+1 for i in range(len(df['id']))]
df['id']=np.where(df.id>-1,df.id+1,0)
df['links']=np.where(df.links>=0,df.links+1,0)
df.head(10)
```

```python
import copy


template = {
    "id": 0,
    "text": [],
    "box": [],
    "words":[]
}
output={
    "form": []
}
# function to get data of label
def getLabel(row):
    label=copy.deepcopy(template)
    # set the label id
    label['id']=row['id']
    # set the label text
    label['text']=row['word']
    # set the label box
    label['box']=[row['word_x1'],row['word_y1'],row['word_x2'],row['word_y2']]
    return label
def getWord(row):
    label=copy.deepcopy(template)
    del label['words']
    del label['id']
    # set the label text
    label['text']=row['word']
    # set the label box
    label['box']=[row['word_x1'],row['word_y1'],row['word_x2'],row['word_y2']]
    return label
```

```python
idDict={}
n=0
# iterate over the dataframe
for index, row in df.iterrows():
    # create a new label
    if not row['links']:
        word=getWord(row)
        label=getLabel(row)
        label['words'].append(word)
        # add the label to the output
        output['form'].append(label)
        idDict[row['id']]=n
        n+=1
    # increment the label id
    # ids+=1
output
```
